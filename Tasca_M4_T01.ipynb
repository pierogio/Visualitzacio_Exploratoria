### Limpieza

#### Al parecer el dtype de la columna "Preu" es de tipo "object" cuando debería ser de tipo numérico. Además, la columna misma incluye unos valores incorrectos identificados como "--".
#### Como tener valores a cero o hacer una media influiria en el análisis y realmente no tenemos datos, la opción más sensata parece eliminar las rows que contienen dichos valores.
#### Además, hay columnas de las cuales se puede prescindir como "Any" (solo tenemos datos del mismo año) y "Codi_Districte”, "Codi_Barri" (tenemos los nombres de los mismos).

```python
lloguer_barri=lloguer_barri.drop(['Any', 'Codi_Districte', 'Codi_Barri'], axis=1)
```

```python
lloguer_barri.loc[lloguer_barri['Preu'] == '--', 'Preu'] = np.nan
```

```python
lloguer_barri.isnull().sum()
```

```python
lloguer_barri['Preu']=lloguer_barri['Preu'].astype("float")
```

```python
lloguer_barri.info()
```

```python
lloguer_barri=lloguer_barri.dropna()
```

```python
#Visualizo el resultado tras eliminar columnas y NaN. 
msno.bar(lloguer_barri)
plt.show()
```

### Visualización Grafica

```python
#Visualización de una variable categorica mostrando el porcentaje de la distribución de barrios por cada distrito (en base a las cantidades cuyos datos disponemos):
plt.style.use('_mpl-gallery-nogrid')

lloguer_barri[['Nom_Districte','Nom_Barri']].sort_values(by=['Nom_Districte'],ascending=False).drop_duplicates().Nom_Districte.value_counts().plot(kind='pie', ylabel='', autopct='%1.1f%%', figsize=(8, 8), 
        wedgeprops = {"edgecolor" : "white",
                      'linewidth': 2,
                      'antialiased': True})

plt.title("Distribución en porcentaje de barrios por cada districto")
plt.legend(bbox_to_anchor=(1.2, 1))
#guardo imagen del grafico
plt.savefig("Porcentaje.png", bbox_inches='tight', dpi=200)
```

```python
#Visualización de una variable categorica mosntrando la cantidad de barrios con datos disponibles por cada distrito.
plt.style.use('default')
lloguer_barri[['Nom_Districte','Nom_Barri']].sort_values(by=['Nom_Districte'],ascending=False).drop_duplicates().Nom_Districte.value_counts().plot(kind="bar", figsize=(8,4))
plt.title("Cantidad de barrios disponibles por cada districto")
plt.xlabel("Distrito")
plt.xticks(rotation=45)
plt.ylabel("Cantidad")
plt.show()
```

##### Podemos observar que los distritos de San martí, Nou Barris y Horta-Guinardó son los districos que disponen de mas barrios y datos.

```python
#Visualizo de una variable numerica mostrando la fequencia del rango de los valores por precio mensual.

sns.set(rc={"figure.figsize":(10, 6)})
sns.set_style(rc = {'axes.facecolor': 'white'})

sns.histplot(lloguer_barri.loc[lloguer_barri['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)']['Preu'], color='C0')
plt.title("Rango de precios")
plt.xlabel("Preu")
plt.xticks(rotation=45)
plt.ylabel("Cantidad")
#guardo imagen del grafico
plt.savefig("Rango.png", bbox_inches='tight', dpi=200)
```

##### Podemos observar que el rango de precios está alrededor de los 800€.

```python
#Visualización de una variable numerica y una categorica mostrando la distribución de los precios mensuales por distrito
y = "Nom_Districte"
x = "Preu"

fig, axs = plt.subplots(ncols=2, figsize=(15, 10))

plot1=sns.boxplot(data=lloguer_barri.loc[lloguer_barri['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)'], y=x, x=y, ax=axs[0])
plot1.set_xticklabels(plot1.get_xticklabels(), rotation=90)

plot2=sns.stripplot(y= x,data=lloguer_barri.loc[lloguer_barri['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)'], x =y, hue="Nom_Districte", ax=axs[1])
plot2.set_xticklabels(plot2.get_xticklabels(), rotation=90)


fig.tight_layout()
#guardo imagen del grafico
plt.savefig("Distribucion.png", bbox_inches='tight', dpi=200)
```

```python
#Visualización de dos variables numericas mostrando la distribución de precios por m2 para cada trimestre.
plt.figure(figsize=(10,8))
sns.boxplot(x = "Trimestre",
            y = "Preu",
            data = lloguer_barri.loc[lloguer_barri['Lloguer_mitja']=='Lloguer mitjà per superfície (Euros/m2 mes)'])
sns.stripplot(x = "Trimestre",
              y = "Preu",
              color = 'black',
              alpha = 0.3,
              data = lloguer_barri.loc[lloguer_barri['Lloguer_mitja']=='Lloguer mitjà per superfície (Euros/m2 mes)'])
plt.title("Distribución m2/Trimestre", size=18)

plt.show()
```

#### El análisis de los gráficos de caja y bigotes parece indicar que hay una leve tendencia de aumento en los precios por metro cuadrado a medida que avanzan los trimestres. La mayoría de los datos se mantiene en un rango relativamente estrecho y al parecer hay pocos valores atípicos o inusuales.

```python
#Visualización de tres variables, districto, trimestre y precio, haciendo la media de los precios por cada distrito en cada trimestre.
sns.barplot(data=round(lloguer_barri.loc[lloguer_barri['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)'][['Nom_Districte','Trimestre','Preu']].groupby(['Nom_Districte', 'Trimestre']).mean().reset_index(),2), x="Trimestre", y="Preu", hue="Nom_Districte")
plt.legend(bbox_to_anchor=(1.0, 1))

plt.show()
```

#### En resumen, parece ser que los distritos mas caros son Sarrià y Les Corts, mientras que Nou Barris se mantiene como barrio con el precio medio mas bajo durante todo el año.
